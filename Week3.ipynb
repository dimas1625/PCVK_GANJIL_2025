# Linear Brigthness Transformation result

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# Pastikan path sesuai file upload
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

# Cek apakah gambar terbaca
if img is None:
    print("Gambar tidak ditemukan, cek path filenya!")
else:
    # Konversi ke RGB
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # Input nilai kecerahan
    brightness = int(input("Masukkan nilai kecerahan: "))

    # Linear Brightness Transformation
    bright_img = cv.convertScaleAbs(img_rgb, alpha=1, beta=brightness)

    # Tampilkan hasil
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img_rgb)
    plt.title("Citra Asli")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(bright_img)
    plt.title(f"Citra Brightness {brightness:+}")
    plt.axis("off")

    plt.show()
```

#Inverse Transformation

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load gambar dari Google Drive
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

# Pastikan gambar terbaca
if img is None:
    print("⚠️ Gambar tidak ditemukan, cek path filenya!")
else:
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # --- 1. Inverse Transformation ---
    inverse_img = 255 - img_rgb

    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis("off")
    plt.subplot(1,2,2); plt.imshow(inverse_img); plt.title("Inverse Transformation"); plt.axis("off")
    plt.show()
```

#Contrast Image Transformation result

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load gambar dari Google Drive
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

# Pastikan gambar terbaca
if img is None:
    print("⚠️ Gambar tidak ditemukan, cek path filenya!")
else:
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # --- 2. Contrast Transformation ---
    brightness = int(input("Masukkan tingkat kecerahan [-255 .. 255]: "))
    contrast   = float(input("Masukkan kontras [1.0 .. 3.0]: "))

    contrast_img = cv.convertScaleAbs(img_rgb, alpha=contrast, beta=brightness)

    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis("off")
    plt.subplot(1,2,2); plt.imshow(contrast_img); plt.title(f"Contrast α={contrast}, β={brightness}"); plt.axis("off")
    plt.show()
```

#Log Transformation result

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load gambar dari Google Drive
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

# Pastikan gambar terbaca
if img is None:
    print("⚠️ Gambar tidak ditemukan, cek path filenya!")
else:
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # --- 3. Logarithmic Transformation ---
    c = int(input("Masukkan nilai kecerahan (c): "))

    log_img = c * (np.log1p(img_rgb.astype(np.float32)))
    log_img = np.clip(log_img, 0, 255).astype(np.uint8)

    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis("off")
    plt.subplot(1,2,2); plt.imshow(log_img); plt.title(f"Log Transform (c={c})"); plt.axis("off")
    plt.show()
```

#Grayscale Image Averaging method result

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load gambar kucing dari Google Drive
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

if img is None:
    print("⚠️ Gambar tidak ditemukan, cek path!")
else:
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # --- a. Averaging method ---
    avg_gray = np.mean(img_rgb, axis=2).astype(np.uint8)

    # Plot hasil
    plt.figure(figsize=(15,5))

    plt.subplot(1,4,1)
    plt.imshow(img_rgb)
    plt.title("Original")

    plt.subplot(1,4,2)
    plt.imshow(avg_gray, cmap="gray")
    plt.title("Averaging")
    plt.show()
```

#Grayscale Image Lightness method result

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load gambar kucing dari Google Drive
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

if img is None:
    print("⚠️ Gambar tidak ditemukan, cek path!")
else:
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # --- b. Lightness method ---
    light_gray = ((np.max(img_rgb, axis=2) + np.min(img_rgb, axis=2)) / 2).astype(np.uint8)

    # Plot hasil
    plt.figure(figsize=(15,5))

    plt.subplot(1,4,1)
    plt.imshow(img_rgb)
    plt.title("Original")

    plt.subplot(1,4,2)
    plt.imshow(light_gray, cmap="gray")
    plt.title("Lightness")
    plt.show()
```

#Grayscale Image Luminance method result

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load gambar kucing dari Google Drive
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

if img is None:
    print("⚠️ Gambar tidak ditemukan, cek path!")
else:
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # --- c. Luminance method ---
    luminance_gray = (0.299*img_rgb[:,:,0] + 0.587*img_rgb[:,:,1] + 0.114*img_rgb[:,:,2]).astype(np.uint8)

    # Plot hasil
    plt.figure(figsize=(15,5))

    plt.subplot(1,4,1)
    plt.imshow(img_rgb)
    plt.title("Original")

    plt.subplot(1,4,2)
    plt.imshow(luminance_gray, cmap="gray")
    plt.title("Luminance")
    plt.show()
```

# Image transform with custom color selection

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load gambar dari Google Drive
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

if img is None:
    print("⚠️ Gambar tidak ditemukan, cek path!")
else:
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # Konversi ke HSV
    hsv = cv.cvtColor(img_rgb, cv.COLOR_RGB2HSV)

    # Range warna biru (bisa disesuaikan)
    lower_blue = np.array([100, 50, 50])
    upper_blue = np.array([140, 255, 255])

    # Mask untuk biru
    mask = cv.inRange(hsv, lower_blue, upper_blue)

    # Grayscale untuk bagian non-biru
    gray = cv.cvtColor(img_rgb, cv.COLOR_RGB2GRAY)
    gray_3ch = cv.cvtColor(gray, cv.COLOR_GRAY2RGB)

    # Gabungkan
    final_img = np.where(mask[:,:,None].astype(bool), img_rgb, gray_3ch)

    # Plot hasil
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img_rgb)
    plt.title("Original")

    plt.subplot(1,2,2)
    plt.imshow(final_img)
    plt.title("Blue Highlighted")

    plt.show()
```
