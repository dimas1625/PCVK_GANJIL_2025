```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# Pastikan path sesuai file upload
img = cv.imread("/content/drive/MyDrive/cat1.jpg")

# Cek apakah gambar terbaca
if img is None:
    print("Gambar tidak ditemukan, cek path filenya!")
else:
    # Konversi ke RGB
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # Input nilai kecerahan
    brightness = int(input("Masukkan nilai kecerahan: "))

    # Linear Brightness Transformation
    bright_img = cv.convertScaleAbs(img_rgb, alpha=1, beta=brightness)

    # Tampilkan hasil
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img_rgb)
    plt.title("Citra Asli")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(bright_img)
    plt.title(f"Citra Brightness {brightness:+}")
    plt.axis("off")

    plt.show()
```
