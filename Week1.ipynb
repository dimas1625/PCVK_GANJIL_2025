```python
import cv2 as cv
import numpy as np
import requests
from google.colab.patches import cv2_imshow

# Daftar URL gambar (langsung direct ke .jpg)
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://cdn1.unitedinfocus.com/uploads/14/2024/02/GettyImages-1358485714-1140x757.jpg",  # Ronaldo dari Bing
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg"
]

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

for url in urls:
    try:
        # Ambil gambar dengan headers
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        # Convert ke numpy array
        image_arr = np.asarray(bytearray(response.content), dtype=np.uint8)
        image = cv.imdecode(image_arr, cv.IMREAD_COLOR)

        # Resize & ubah warna
        image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
        image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

        # Gabungkan & tampilkan
        final_frame = cv.hconcat((image, image_2))
        cv2_imshow(final_frame)
        print("\n")
    except Exception as e:
        print(f"Gagal load {url}: {e}")
```

##LANGKAH 3

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = " ,tinggi, " x " ,lebar)
cv2_imshow(image_2)
```

##LANGKAH 4

```python
image_3 = image_2.copy()
for y in range(lebar):
    image_3[int(tinggi/2), y] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

###2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
# Garis vertikal di tengah
image_4 = image_2.copy()
for x in range(tinggi):
    image_4[x, int(lebar/2)] = [255, 255, 255]

# Garis diagonal dari kiri atas ke kanan bawah
image_5 = image_2.copy()
for i in range(min(tinggi, lebar)):
    image_5[i, i] = [255, 255, 255]
for i in range(min(tinggi, lebar)):
    image_5[i, lebar - 1 - i] = [255, 255, 255]

final_frame = cv.hconcat([image_4, image_5])
cv2_imshow(final_frame)
```

```python
image_6 = image_2.copy()
start_y = int(lebar * 0.25)
end_y = int(lebar * 0.75)
for y in range(start_y, end_y):
    image_6[int(tinggi/2), y] = [255, 255, 255]

cv2_imshow(image_6)
```

```python
image_7 = image_2.copy()
start_x, start_y = 100, 150
size = 100
for x in range(start_x, start_x + size):
    for y in range(start_y, start_y + size):
        image_7[x, y] = [255, 255, 255]

cv2_imshow(image_7)
```
